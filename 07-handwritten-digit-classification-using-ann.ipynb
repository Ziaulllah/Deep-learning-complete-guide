```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_test.shape
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[2])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

# ✅ Path to your uploaded image in the Kaggle dataset
img_path = "/kaggle/input/hw-image/HW image.png"

# === Load and Process the Image ===
image = Image.open(img_path).convert('L')         # Convert to grayscale
image = ImageOps.invert(image)                    # Invert if digit is black on white
image = image.resize((28, 28))                    # Resize to 28x28
img_array = np.array(image) / 255.0               # Normalize (0 to 1)
img_array = img_array.reshape(1, 28, 28)           # Reshape for model input

# === Predict Using the Trained Model ===
prediction = model.predict(img_array)
predicted_digit = prediction.argmax(axis=1)[0]

# === Display the Result ===
print("✅ Predicted Digit:", predicted_digit)
plt.imshow(img_array.reshape(28, 28), cmap='gray')
plt.title(f"Predicted: {predicted_digit}")
plt.axis('off')
plt.show()
```
